```python
#$ Task: (Reshape the US dataset from HW1)
path= 'C:/Users/USER/Desktop/homeworkhw0/accident_all_years'
accidents_all_years=glob.glob(path + "/*.csv")
accidents_all_years
li=[]
for filename in accidents_all_years:
    df=pd.read_csv(filename, index_col=None, header=0,encoding='latin-1')
    li.append(df)
li
df=pd.concat (li, axis=0, ignore_index=True, sort=False)
df.head()
```

```python
df.columns
```

```python
df1=pd.DataFrame(df, columns=['PERSONS', 'FATALS' ])
df1.head()
                                
```

```python
df2=df1.dropna(how='any')
df2.shape
```

```python
# Making a copy of the dataframe
df3 = df2.copy()
scaler = StandardScaler()
scaled_cols = ["PERSONS", "FATALS"]
df3[scaled_cols] = scaler.fit_transform(df2[scaled_cols])
df3.head()

```

```python
# Task: (Perform ad hoc K-Means clustering)
kmeansDt = df3[["PERSONS", "FATALS"]]
kmeansDt.head()
```

```python
Fvalues = kmeansDt.values
Fvalues
```

```python
#Grouping the data with 10 cluster values
kmeans = KMeans(n_clusters=10, random_state=100)
kmeans.fit(Fvalues)
y_kmeans = kmeans.predict(Fvalues)
print(y_kmeans)
```

```python
##Grouping the data with 5 cluster values
kmeans = KMeans(n_clusters=5, random_state=100)
kmeans.fit(Fvalues)
y_kmeans = kmeans.predict(Fvalues)
print(y_kmeans)
```

```python
#Grouping the data with 3 cluster values
kmeans = KMeans(n_clusters=3, random_state=100)
kmeans.fit(Fvalues)
y_kmeans = kmeans.predict(Fvalues)
print(y_kmeans)
```

```python
#Grouping the data with 2 cluster values
kmeans = KMeans(n_clusters=2, random_state=100)
kmeans.fit(Fvalues)
y_kmeans = kmeans.predict(Fvalues)
print(y_kmeans)
```

```python
#$ Task : Perform elbow analysis to find optimal cluster size
css= []
max_clusters =15
for k in range(1,max_clusters):
 kmeans = KMeans(init='k-means++',max_iter=200,random_state=0, n_clusters=k)
 kmeans.fit_predict(df3)
 css.append(kmeans.inertia_)
 
plt.plot(range(1,len(css)), css[1:])
plt.xlabel ('No of clusters')
plt.ylabel('css')
```
